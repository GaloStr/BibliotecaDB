```python
from config import get_db_connection, pd, plt, sns, MaxNLocator
conn = get_db_connection()
```

```python
sql_top_libros = """
SELECT titulo,autor, total_prestamos
FROM Libros_Mas_Prestados
ORDER BY total_prestamos DESC
"""
```

```python
df_top_libros = pd.read_sql_query(sql_top_libros, conn)
conn.close()
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(16, 5))
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
sns.barplot(x='titulo', y='total_prestamos', data=df_top_libros, palette='plasma') 
plt.title('Libros Más Prestados (Conteo Total)')
plt.xlabel('Número de Préstamos') 
plt.ylabel('Título del Libro')
plt.show()
```
