```python
from config import get_db_connection, pd, plt, sns 
conn = get_db_connection()
```

```python
sql_top_ciudades = """
SELECT ciudad, total_prestamos
FROM KPI_Ventas_Por_Ciudad
ORDER BY total_prestamos DESC;
"""
```

```python
df_top_ciudad = pd.read_sql_query(sql_top_ciudades, conn)
conn.close()
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(6, 6))
sns.barplot(x='ciudad', y='total_prestamos', data=df_top_ciudad, palette='plasma')

plt.title('KPI: Rendimiento de Préstamos por Ciudad', fontsize=16)
plt.xlabel('Ciudad de Residencia del Socio', fontsize=12)
plt.ylabel('Número Total de Préstamos', fontsize=12)

plt.show()
```
